## 4 | Data Pre-Processing

```python
# Define functions for data pre-processing

def normalizeTrain(df):
  scaler = StandardScaler()
  array = scaler.fit_transform(df)

  return pd.DataFrame(array, columns=df.columns), scaler

def normalizeTest(df, scaler):
  array = scaler.transform(df)
  return pd.DataFrame(array, columns=df.columns)

def split_features_and_target(df):
  x = df.drop(['target'],axis=1)
  y = df[['target']]
  return x, y

def split_df(df):
  x, y = split_features_and_target(df)
  x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=1,test_size=0.2)
  return x_train, x_test, y_train, y_test

def process_train_data(x_train):
  x_train, scaler_x = normalizeTrain(x_train)
  return x_train, scaler_x

def process_test_data(x_test, scaler_x):
  x_test = normalizeTest(x_test, scaler_x)
  return x_test

def negative_to_zero(df):
  df[df < 0] = 0
  return df

def map_target_to_integers(df):
  bacteria_integers = {"target": {"Bacteroides_fragilis":1,
                    "Streptococcus_pyogenes":2,
                    "Streptococcus_pneumoniae":3,
                    "Campylobacter_jejuni":4,
                    "Salmonella_enterica":5,
                    "Escherichia_coli":6,
                    "Enterococcus_hirae":7,
                    "Escherichia_fergusonii":8,
                    "Staphylococcus_aureus":9,
                    "Klebsiella_pneumoniae":10}}
  df = df.replace(bacteria_integers)
  return df
```

```python
# Map target bacterias to integers
train_data = map_target_to_integers(train_data)
```

```python
# Replace negative values with 0 as they are assumed to be insignificant for training
train_data = negative_to_zero(train_data)
```

```python
# Split data into test_data and train_data
x_train, x_test, y_train, y_test = split_df(train_data)
```

```python
# Add standard scaling to dataset
x_train, scaler_x = process_train_data(x_train)
x_test = process_test_data(x_test, scaler_x)
```

## 5 | Model Testing

```python
# Import Models
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import SGDClassifier
```

```python
# Initialize Models and Parameters
mlp = MLPClassifier(hidden_layer_sizes=(100,),activation='tanh',early_stopping=True, random_state=3)
rfc = RandomForestClassifier(max_depth=2, random_state=0)
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
sgd = SGDClassifier(max_iter=1000, tol=1e-3)
```

```python
# Test for Multi Layer Perceptron Classifier
mlp.fit(x_train, y_train)
score = mlp.score(x_test, y_test)
print(score)
```

```python
# Test for Random Forest Classifier
rfc.fit(x_train, y_train)
score = rfc.score(x_test, y_test)
print(score)

# https://scikit-learn.org/stable/modules/model_persistence.html
# dump(mlp, 'one_cycle.joblib')
```

```python
# Test for Gradient Boosting Classifier
gbc.fit(x_train, y_train)
score = gbc.score(x_test, y_test)
print(score)
```

```python
# Test for Stochastic Gradient's Descent Classifier
sgd.fit(x_train, y_train)
score = sgd.score(x_test, y_test)
print(score)
```

## 6 | Submission

```python
test_data_raw = pd.read_csv("test.csv", index_col=False)
row_ids = test_data_raw['row_id']
test_data = test_data_raw.drop(['row_id'],axis=1)
test_data = map_target_to_integers(test_data)
test_data = negative_to_zero(test_data)
x_final = process_test_data(test_data, scaler_x)
y_predicted = mlp.predict(x_final)
integers_bacteria = {"target": {1:"Bacteroides_fragilis",
                2:"Streptococcus_pyogenes",
                3:"Streptococcus_pneumoniae",
                4:"Campylobacter_jejuni",
                5:"Salmonella_enterica",
                6:"Escherichia_coli",
                7:"Enterococcus_hirae",
                8:"Escherichia_fergusonii",
                9:"Staphylococcus_aureus",
                10:"Klebsiella_pneumoniae"}}
y_predicted = pd.DataFrame(y_predicted, columns = ['target'])
y_predicted = y_predicted.replace(integers_bacteria)

submission_df = row_ids.concat(y_predicted)
print(submission_df)
```

```python
submission_df.to_csv("./submission.csv")
```
